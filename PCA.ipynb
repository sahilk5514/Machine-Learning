```python
import  matplotlib.pyplot as plt
import seaborn as sns
import numpy as np 
import pandas as pd
%matplotlib inline 
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer_dataset = load_breast_cancer()
```

```python
print(cancer_dataset.DESCR)
```

```python
df =pd.DataFrame(cancer_dataset["data"],columns=cancer_dataset['feature_names'])
```

```python
df.head()
```

```python
df.columns
```

```python
## Standardization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit(df)
```

```python
data_pca = scaler.transform(df)
```

```python
data_pca
```

```python
## Applying PCA algorithms
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=5)   #transformed all columns into 2 columns
data_pca= pca.fit_transform(data_pca)
```

```python
data_pca
```

```python
pca.explained_variance_
```

```python
plt.figure(figsize=(8,6))
plt.scatter(data_pca[:,0],data_pca[:,1],c=cancer_dataset["target"],cmap="plasma")
plt.xlabel('First principal Component')
plt.ylabel('Second principal Component')
```
